# **House pricedata**

```python
import numpy as np
import pandas as pd
```

```python
data=pd.read_csv("/content/homeprices.csv")
```

```python
dummies=pd.get_dummies(data["town"]).astype(int)
dummies
```

```python
merge=pd.concat([data,dummies],axis=1)
merge
```

```python
data.drop(["town"],axis="columns",inplace=True)
```

```python
merge=merge.drop(["town"],axis=1)
```

```python
y=merge["price"]
x=merge.drop(["price"],axis=1)
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()

model.fit(x,y)
```

```python
x
```

```python
model.predict([[2600,1,0,0]])
```

# **prediction method 2**

```python
import numpy as np
import pandas as pd
```

```python
data=pd.read_csv("/content/homeprices.csv")
```

```python
from sklearn.preprocessing import OneHotEncoder
encoder = OneHotEncoder(sparse_output=False)
encoder
```

```python
one_hot_encoded = encoder.fit_transform(data[["town"]])
one_hot_data = pd.DataFrame(one_hot_encoded,
columns=encoder.get_feature_names_out(input_features=["town"]))
one_hot_data
```

```python
data_encoded=pd.concat([data,one_hot_data],axis=1)
data_encoded
```

```python
data_encoded=data_encoded.drop("town",axis=1)
data_encoded
```

```python
from sklearn.linear_model import LinearRegression
y=data_encoded["price"]
x=data_encoded.drop(["price"],axis=1)
model=LinearRegression()
model.fit(x,y)
```

```python
model.predict([[2600,1,0,0]])
```
